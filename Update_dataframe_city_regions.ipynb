#  Set-up

```python
# import libraries
import pandas as pd
```

# Import last version of excel

```python
#Retrive cities list
data = pd.read_excel('/content/drive/MyDrive/ItalianCitiesV15.xlsx')
data
```

```python
data = data.drop('Unnamed: 0',axis=1)
data
```

# Check if a value exist or not

```python
# Check if the value exists in any column of the DataFrame

value_to_check = 'Giambellino'

value_exists = data.isin([value_to_check]).any().any()

if value_exists:
    print(f"The value '{value_to_check}' exists in the DataFrame.")
else:
    print(f"The value '{value_to_check}' does not exist in the DataFrame.")
```

# Create a new row (city and region)

```python
# Create a new row
city = 'Giampellino'
region = 'Lombardia'
new_row_data = {'city_ita': city, 'region_ita': region}

# Use loc to add the new row to the DataFrame
data.loc[len(data)] = new_row_data
data
```

# Update an existing row (city and region)

```python
# Update a row
old_city = 'brindisi'
old_region = 'Puglia'
new_city = 'Brindisi'
new_region = 'Puglia'

update_data = {'city_ita': new_city, 'region_ita': new_region}

print(f'Do you want to update {old_city} and {old_region} with {new_city} and {new_region}?')

confirmation = input('If yes, press y and ENTER; if not, any other key and ENTER: ')

if confirmation.lower() == 'y':
    # Use boolean indexing to update the row
    data.loc[data.index[data['city_ita'] == old_city].tolist()[0]] = update_data
    print("DataFrame updated successfully.")
    print(data.loc[data.index[data['city_ita'] == new_city].tolist()[0]])
else:
    print("No changes were made.")
```

```python
# Check if operation it's fine

check = data[data['city_ita'] == new_city]
print(check)
```

# Export excel to new version

```python
excel_file_path = '/content/drive/MyDrive/ItalianCitiesV16.xlsx'
data.to_excel(excel_file_path,'Italian cities and regions')
print(f"DataFrame has been exported to {excel_file_path}")
```
