```python
import pandas as pd
import numpy as np
import time
# import package that remove all accent from a word
from unidecode import unidecode
```

# Italian cities list

```python
# Retrive excel with all cities and relative regions
# If you want to update city and regions in file, please refer to script "Update_dataframe_city_regions"
excel_google_drive_path = '/content/drive/MyDrive/ItalianCitiesV16.xlsx'
cities = pd.read_excel(excel_google_drive_path)
cities
```

# Import Keywords

```python
# Retrive excel with all keywords
keyword_path = '/content/drive/MyDrive/unipolsai.it-organic.Positions-it-20240112-2024-01-13T11_09_54Z.xlsx'
data = pd.read_excel(keyword_path)
data
```

```python
short_data = data[['Keyword']].copy()
short_data
```

# Cluster

```python
# insert in this list all words that can be referred to a street and not to a city
street_word = ['via','piazza','viale','corso','piazzale']
```

```python
start_time = time.time()

def cities_regions(keywords):

    # split keyword in a list o words
    split = keywords.split()

    # remove accent to every word
    split = [unidecode(s) for s in split]

    # drop follow word if in street_word list exist (via roma , word roma will be delete)
    index_to_remove = -1  # Initialize with an invalid index
    for index,word in enumerate(split):
        if word in street_word:
          index_to_remove = index + 1
    try:
      if index_to_remove != -1:
        split.pop(index_to_remove)
    except:
      pass

    # check if any city it's inside list of keywords
    matching_cities = cities[cities['city_ita'].apply(lambda x: all(unidecode(city.lower()) in split for city in x.split()))]

    # compile dataframe column
    if not matching_cities.empty:
        return matching_cities['city_ita'].iloc[0]
    else:
        return np.nan

def city_to_region(city):

    # write the region name if city is founded
    matching_region = cities.loc[cities['city_ita'] == city, 'region_ita']

    if not matching_region.empty:
        return matching_region.iloc[0]
    else:
        return np.nan

# Apply the functions to create new columns
short_data['Sottocluster_Semantico_(Citta)'] = short_data['Keyword'].apply(cities_regions)
short_data['Cluster_Semantico_(Regione)'] = short_data['Sottocluster_Semantico_(Citta)'].apply(city_to_region)
end_time = time.time()
print(f'Time of excecution: {round(end_time-start_time,2)}')
```

# Check

```python
# Check list of cities not founded in keyword
check_df = short_data[pd.isna(short_data['Sottocluster_Semantico_(Citta)'])]
# Calculate the percentage of success city check
nan_percentage = short_data['Sottocluster_Semantico_(Citta)'].isna().mean() * 100
print(f'Match values for keyword: {str(round(100-nan_percentage,2))} % (Number of not matches: {round(nan_percentage,2)}%)')
print(f'Number of values without cluster: {check_df.shape[0]}')
print('Keywords not clustered:')
check_df
```

```python
check_kw = short_data[pd.notna(short_data['Sottocluster_Semantico_(Citta)'])]
check_kw
```
